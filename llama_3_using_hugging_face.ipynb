````python
import json
import torch
from transformers import (AutoTokenizer,
                          AutoModelForCausalLM,
                          BitsAndBytesConfig,
                          pipeline)
````

**HF account Configuration**

this needs a config.json that looks


```
{
  "HF_TOKEN":your token here
}
```

````python
config_data = json.load(open("config.json"))
HF_TOKEN = config_data["HF_TOKEN"]
````

````python
model_name = "meta-llama/Meta-Llama-3-8B"
````

**Quantisation Configuration**

This shrinks the data, under the max of colab. Remove this for your own computer

````python
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.bfloat16
)
````

**Loading the Tokenizer and the LLM**

````python
tokenizer = AutoTokenizer.from_pretrained(model_name,
                                          token=HF_TOKEN)

tokenizer.pad_token = tokenizer.eos_token
````

````python
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    device_map="auto",
    quantization_config=bnb_config,
    token=HF_TOKEN
)
````

````python
text_generator = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    max_new_tokens=128
)
````

````python
def get_response(prompt):
  sequences = text_generator(prompt)
  gen_text = sequences[0]["generated_text"]
  return gen_text[len(prompt):]
````

````python
prompt = "What is Machine Learning?"
````

````python
llama3_response = get_response(prompt)
llama3_response
````
